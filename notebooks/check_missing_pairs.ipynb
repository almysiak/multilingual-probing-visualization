```python
import pandas as pd
from itertools import product
import os
```

```python
LANGS = ('be', 'cs', 'de', 'en', 'es', 'fi', 'fr', 'id', 'lv', 'lt', 'pl', 'ru', 'slk', 'slv', 'uk', 'zh', 'ar', 'bg', 'tr', 'ko', 'hy', 'fa', 'ja', 'ga', 'cy', 'hr', 'nl')
```

```python
len(LANGS)
```

```python
max_dict = {
    'ar': 6074,
    'pl': 17721,
    'bg': 8906,
    'be': 22852,
    'fr': 14448,
    'en': 12542,
    'lv': 12520,
    'fa': 4797,
    'lt': 2340,
    'cs': 68494,
    'es': 14286, 
    'zh': 3996,
    'slk': 8482,
    'slv': 10902,
    'uk': 5495,
    'ru': 69629,
    'de': 13813,
    'fi': 12216,
    'id': 4481,
    'hr': 6913,
    'tr': 14850,
    'ko': 23010,
    'hy': 1974,
    'fa': 26196,
    'ja': 7050,
    'ga': 4005,
    'cy': 1111,
    'nl': 12289,
}
```

```python
LIMITS = (100, 1000, 2500, 5000, 7500, 10000) # , None)
```

```python
res_df = pd.read_csv('multilingual-probing-visualization/probingOutputs/all_slavic/all_results.csv')
```

```python
res_df['train'] = res_df.train.str[2:-2]
res_df['test'] = res_df.test.str[2:-2]
```

```python
res_df.head()
```

```python
missing = set()
```

```python
counter = 0
for l1, l2, limit in product(LANGS, LANGS, LIMITS):
    if limit is None or max_dict[l2] > int(limit):
        df = res_df[(res_df.test == l1) & (res_df.train == l2) & (res_df.limit == limit)]
    else:
        continue
    if len(df) < 1:
        print(f"Missing: test {l1}, train {l2}, limit {limit}")
        counter += 1
        missing.add((l1, l2, str(limit)))
```

```python
print(f"Missing {counter} exps of {len(set(product(LANGS, LANGS, LIMITS)))}")
```

```python
for filename in os.listdir('multilingual-probing-visualization/configs/all_new'): # skip the Nones for now? and 100s?
    ttl = filename[:-5]
    test, train, limit = ttl.split('_')
    if (test, train, limit) in missing:
        missing.remove((test, train, limit))
        if limit not in (None, 100):
            src_path = os.path.join('multilingual-probing-visualization/configs/all_new', filename)
            dest_path = src_path.replace('all_new', 'missing_new')
            os.rename(src_path, dest_path)
```
